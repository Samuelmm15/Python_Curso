### Ejercicio 1
Visualizar la evolución los casos nuevos por millón de España mediante un gráfico de líneas.

```python
# Pasos previos necesarios para la realización de los ejercicios
import pandas as pd

datos = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv') 

datos # Visualización previa del dataframe usado
```

```python
# Visualización únicamente de los datos de España
datos_españa = datos[datos['location'] == 'Spain']
datos_españa
```

```python
# Realización del ejercicio 1 con gráfico de línea
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.size'] = 30.0

x = datos_españa['date']
y = datos_españa['new_cases_per_million']

fig = plt.figure()
ax = fig.add_axes([1, 1, 2, 2]) # main axes

ax.set_xlabel('Fechas') # nombre eje x 
ax.set_ylabel('Cantidad de casos') # nombre eje y
ax.set_title('Nuevos casos por millón') # nombre gráfica

ax.set_yticks([0,1000000, 2000000, 3000000,4000000,5000000])
ax.set_yticklabels(['0','1 mill','2 mill','3 mill','4 mill', '5 mill'])

plt.plot(x, y, lw=3)
plt.show()
fig.savefig("nuevos_casos_millon.png")
```

### Ejercicio 2
Ahora hacer lo mismo pero también para también Reino Unido, Japón, China y Brazil. Graficar todo en la misma gráfica con España. 

Se tiene que ver leyenda, un título de gráfica y acotados (dándoselos por defecto) los números del eje y

```python
# Paso previo al ejercicio 2
datos_japon = datos[datos['location'] == 'Japan']
datos_china = datos[datos['location'] == 'China']
datos_brazil = datos[datos['location'] == 'Brazil']
datos_reinoUnido = datos[datos['location'] == 'United Kingdom']
```

```python
# Ejercicio 2
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.size'] = 20.0

x_es = datos_españa['date']
x_ja = datos_japon['date']
x_chi = datos_china['date']
x_bz = datos_brazil['date']
x_uk = datos_reinoUnido['date']

y_es = datos_españa['new_cases_per_million']
x_ja = datos_japon['new_cases_per_million']
x_chi = datos_china['new_cases_per_million']
x_bz = datos_brazil['new_cases_per_million']
y_uk = datos_reinoUnido['new_cases_per_million']

fig = plt.figure(figsize=(12,14))

ax = fig.add_axes([1, 1, 2, 2]) # main axes

ax.set_xlabel('Fechas') # nombre eje x 
ax.set_ylabel('Cantidad de casos') # nombre eje y
ax.set_title('Nuevos casos por millón') # nombre gráfica

ax.set_xticklabels(['en-20','ab-20','jul-20','oc-20', 
                    'en-21','ab-21','jun-21','no-21'])

# ax.set_yticks([0,1000000, 3000000,5000000,7000000,9000000])
# ax.set_yticklabels(['0','1 mill','3 mill','5 mill','7 mill', '9 mill'])

plt.plot(x_es, y_es, lw=3,label = "España")
plt.plot(x_ja, x_ja, lw=3,label = "Japón")
plt.plot(x_chi, x_chi, lw=3,label = "China")
plt.plot(x_bz, x_bz, lw=3,label = "Brazil")
plt.plot(x_uk, y_uk, lw=3,label = "Reino Unido")
plt.legend()

plt.show()
fig.savefig("casos_por_millon_paises.png") # Calidad resolucion
```

### Ejercicio 3
Ahora hacer lo mismo pero para los casos nuevos con una gráfica de barras pero con Estados Unidos en vez de España

```python
# Realización del ejercicio 3
datos_EEUU = datos[datos['location'] == 'United States']

condicion = datos_EEUU.new_cases_per_million < 0 # Para la mejor visualización de datos que no son incorrectos
columna = "new_cases_per_million"
valor_nuevo = 0

datos_EEUU.loc[condicion, columna] = valor_nuevo

import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.size'] = 20.0

x = datos_EEUU['date']
y = datos_EEUU['new_cases_per_million']

fig = plt.figure(figsize=(12,14))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes

ax.set_xlabel('Cantidad') # nombre eje x 
ax.set_ylabel('Frecuencia') # nombre eje y
ax.set_title('Frecuencia de nuevos casos por millón en Estados unidos') # nombre gráfica

ax.set_xticklabels(['en-20','ab-20','jul-20','oc-20', 
                    'en-21','ab-21','jun-21','no-21'])

plt.bar(x, y)
plt.show()
fig.savefig("nuevos_casos_por_millon_barra.png") # Calidad resolucion
```

### Ejercicio 4
Ahora hacer una gráfica de queso para los casos totales sólo de America del Norte.

```python
# Paso previo al ejercicio
datos_north_america = datos[datos['location'] == 'North America']
per_cont = datos_north_america.groupby('total_cases').sum()
per_cont
```

```python
# Realización del ejercicio 4
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.size'] = 30.0

values = per_cont.total_cases_per_million
colors = ['r', 'g', 'b', 'm', 'c','#B3AA19']
labels = per_cont.index
plt.pie(values, colors= colors, labels=labels)
fig = plt.gcf()
fig.set_size_inches(15,15) # Para cambiar el tamaño

plt.title('Reparto de casos totales por millón en America del Norte')
plt.show()


# Gráficas de queso
fig.savefig("reparto_queso_america_norte.png") # Calidad resolucion
```
