¿Por qué usar programación como Python, y en concreto Pandas, y no directamente una hoja de cálculo?

Es más útil el uso de la programación con Python y en concreto con Pandas, debido a que Pandas proporciona ciertas\

que si se tratara de una hoja de cálculo. Esto último, sería muchísimo más tedioso trabajandocon grandes cantidades de datos, dónde la hoja de cálculo es tan grande\

### Ejercicio 2
¿Qué es Pandas?

Pandas se trata de una librería disponible en Python 3 que contiene una serie de funciones, métodos, objetos, etc, que son muy útiles para trabajar\

### Ejercicio 3
Diferencia entre los métodos **.iloc()** y **.loc()**

El método **.iloc()** sirve para filtrar filas y columnas concretas de un dataframe, pero, haciendo uso para ello\
del índice de estas filas y columnas. Mientras, el método **.loc()** sirve para filtrar filas y columnas, pero en este caso, haciendo uso de las etiquetas de estas.

### Ejercicio 4
¿Para qué se usa groupby?

**groupby** sirve para agrupar ciertas columnas, cuyas claves son pasadas como parámetro del método.
