Ejercicio 6: media y mediana de cc entre los dias 20 y 30 del mes de enero del año 2018 y el máximo

```python
dia = datos_copy[(datos_copy['day'] >= 20) & (datos_copy['day'] <= 30) & (datos_copy['month'] == 1) & (datos_copy['year'] == 2018)]
col_cc = dia['cc']

# Media
print(f'Media: {col_cc.mean()}')
# Mediana
print(f'Mediana: {col_cc.median()}')
# Máximo
print(f'Máximo: {col_cc.max()}')
```

**Métodos de agrupación:**

```python
# Agrupación
media_por_mes = datos_copy.groupby('month').mean()

media_por_mes.loc[:,['eol']]
```

**Esto a continuación sirve para poder establecer una condición en la que se debe generar que dicha celda debe de cumplir la condición especificada,\
con ello se puede modificar aquellas celdas que estén mal.**

```python
condicion = datos_copy.eol < 0       # queremos coger todos los datos de generación 
                                # menores a 0
columna = "eol" 
valor_nuevo = 0 # Valor por el que vamos a sustituir

datos_copy.loc[condicion, columna] = valor_nuevo
```

Se puede realizar el groupby juntando distintas condiciones:

```python
# Agrupamos
media_año_mes = datos.groupby(['month','year']).mean()

# Sólo datos eólicos
media_eol = media_año_mes.loc[:,['eol']]

media_eol.head()
```
